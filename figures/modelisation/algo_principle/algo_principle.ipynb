```python
import os
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial import distance_matrix
from scipy.stats import rankdata
```

```python
FIG_PATH = Path(os.path.abspath(''))
FIG_PATH
```

```python
# Simulate the projection of the songs
X = np.random.normal(0,1, size = 1000)
Y = np.random.normal(0,1, size = 1000)
```

```python
# Simulate a user musical taste
x_user = np.random.normal(-1, .2, size = 50)
y_user = np.random.normal(-1, .2, size = 50)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X,Y, color ='grey', alpha = 0.7, label = 'Tracks')
plt.scatter(x_user,y_user, color ='red', alpha = 0.7, label = 'User playlist' )
plt.legend(loc = "lower right")
plt.savefig(FIG_PATH/f'mono_taste_user_playlist.png')
```

```python
# Compute the user standard profile
x_avg = x_user.mean(axis = 0)
y_avg = y_user.mean(axis = 0)
```

```python
plt.figure(figsize=(5,5))
plt.scatter(X,Y, color ='grey', alpha = 0.7, label = 'Tracks')
plt.scatter(x_user,y_user, color ='red', alpha = 0.7, label = 'User playlist' )
plt.scatter(x_avg,y_avg, marker = 'x', color ='black', alpha = 1,  label = 'User centroid')
plt.legend(loc = "lower right")
plt.savefig(FIG_PATH/f'mono_taste_user_center.png')
```

```python
### find the k nearest neighbors
avg = np.array([[x_avg ,y_avg]])
D = distance_matrix(np.concatenate([[X],[Y]]).transpose(), avg)
x_reco = X[np.where(rankdata(D)<=10)[0]]
y_reco = Y[np.where(rankdata(D)<=10)[0]]
```

```python
plt.figure(figsize=(5,5))

plt.scatter(X,Y, color ='grey', alpha = 0.7, label = 'Tracks')
plt.scatter(x_user,y_user, color ='red', alpha = 0.7, label = 'User playlist' )
plt.scatter(x_reco,y_reco, color ='green', alpha = 1,  label = '10 nearest tracks')
plt.scatter(x_avg,y_avg, marker = 'x', color ='black', alpha = 1,  label = 'User centroid')
plt.legend(loc = "lower right")
plt.savefig(FIG_PATH/f'mono_taste_user_avg_reco.png')
```

```python
# Example of a user with a second favorite musical genre
x2_user = np.random.normal(1, .2, size = 50)
y2_user = np.random.normal(1, .2, size = 50)
```

```python
plt.figure(figsize=(5,5))

plt.scatter(X,Y, color ='grey', alpha = 0.7, label = 'Tracks')
plt.scatter(np.concatenate([x_user, x2_user]),np.concatenate([y_user,y2_user]), color ='red', alpha = 0.7, label = 'User playlist' )
plt.legend(loc = "lower right")
plt.savefig(FIG_PATH/f'bi_taste_user_playlist.png')
```

```python
# Compute the user standard profile
x_avg = np.concatenate([x_user, x2_user]).mean()
y_avg = np.concatenate([y_user,y2_user]).mean()
```

```python
plt.figure(figsize=(5,5))

plt.scatter(X,Y, color ='grey', alpha = 0.7, label = 'Tracks')
plt.scatter(np.concatenate([x_user, x2_user]),np.concatenate([y_user,y2_user]), color ='red', alpha = 0.7, label = 'User playlist' )
plt.scatter(x_avg,y_avg, marker = 'x', color ='black', alpha = 1,  label = 'User centroid')
plt.legend(loc = "lower right")
plt.savefig(FIG_PATH/f'bi_taste_user_avg.png')
```

```python
### find the k nearest neighbors
avg = np.array([[x_avg ,y_avg]])
D = distance_matrix(np.concatenate([[X],[Y]]).transpose(), avg)
x_reco = X[np.where(rankdata(D)<=10)[0]]
y_reco = Y[np.where(rankdata(D)<=10)[0]]
```

```python
plt.figure(figsize=(5,5))

plt.scatter(X,Y, color ='grey', alpha = 0.7, label = 'Tracks')
plt.scatter(np.concatenate([x_user, x2_user]),np.concatenate([y_user,y2_user]), color ='red', alpha = 0.7, label = 'User playlist' )
plt.scatter(x_reco,y_reco, color ='green', alpha = 1,  label = '10 nearest tracks')
plt.scatter(x_avg,y_avg, marker = 'x', color ='black', alpha = 1,  label = 'User centroid')
plt.legend(loc = "lower right")
plt.savefig(FIG_PATH/f'bi_taste_user_avg_reco.png')
```

```python
# Compute the user centroids
x_avg = x_user.mean(axis = 0)
y_avg = y_user.mean(axis = 0)

x2_avg = x2_user.mean(axis = 0)
y2_avg = y2_user.mean(axis = 0)
```

```python
plt.figure(figsize=(5,5))

plt.scatter(X,Y, color ='grey', alpha = 0.7, label = 'Tracks')
plt.scatter(np.concatenate([x_user, x2_user]),np.concatenate([y_user,y2_user]), color ='red', alpha = 0.7, label = 'User playlist' )
plt.scatter(np.concatenate([[x_avg], [x2_avg]]),np.concatenate([[y_avg], [y2_avg]]), marker = 'x', color ='black', alpha = 1,  label = 'User centroids')
plt.legend(loc = "lower right")
plt.savefig(FIG_PATH/f'bi_taste_user_centroids.png')
```

```python
### find the k nearest neighbors
avg = np.array([[x_avg ,y_avg]])
D = distance_matrix(np.concatenate([[X],[Y]]).transpose(), avg)
x_reco = X[np.where(rankdata(D)<=5)[0]]
y_reco = Y[np.where(rankdata(D)<=5)[0]]

avg2 = np.array([[x2_avg ,y2_avg]])
D2 = distance_matrix(np.concatenate([[X],[Y]]).transpose(), avg2)
x2_reco = X[np.where(rankdata(D2)<=5)[0]]
y2_reco = Y[np.where(rankdata(D2)<=5)[0]]
```
