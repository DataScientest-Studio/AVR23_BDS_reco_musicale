```python
import os
from pathlib import Path
import pandas as pd

CUR_DIR = os.path.abspath('')
DATA_DIR = Path(CUR_DIR).parent / "data" / "millionsong"
```

Load the second file track_metadata.db containing tracks informations

```python
CUR_DIR = DATA_DIR / 'track_metadata.db'
import numpy as np

import sqlite3
from pandas import read_sql_query, read_sql_table
with sqlite3.connect(CUR_DIR) as dbcon:
    tables = list(read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", dbcon)['name'])
    df_metadata = {tbl : read_sql_query(f"SELECT * from {tbl}", dbcon) for tbl in tables}
    #print(df_metadata)
print(df_metadata.keys())
print(df_metadata.items())
df_metadata = pd.DataFrame.from_dict(df_metadata['songs'])
```

```python
df_metadata
```

```python
df_metadata['song_id'].isin(df_triplets['track_id']).sum()
```

Rename df_triplets track_id key into song_id key to merge the two dataframes

```python
df_triplets = df_triplets.rename(columns = {'track_id' : 'song_id'})
```

```python
print(df_metadata['song_id'].duplicated().sum())
df_metadata[df_metadata['song_id'].duplicated()]
```

In the metadata dataframe, some songs are duplicates

```python
df_metadata = df_metadata.drop_duplicates(subset = ['song_id'], keep='first')
df_metadata['song_id'].duplicated().sum()
```

Now let's merge the two dataframes on the common key song_id

```python
df_triplets_metadata = pd.merge(df_triplets, df_metadata, on = 'song_id', how = 'inner')
df_triplets_metadata
```

```python
## Suppression of cold_starters by selecting 1e+5 and 1e+4 largest listeners
## results in dataframes of respectively  1.771.670  and 290.274 rows with a minimal listening counts of 35 and 116 


## first 1e+5 "extreme users" who can listen to a single song more than a thousand times
df_filter_1 = df_triplets_metadata_spotify.groupby('user')\
                                        .agg({'listening_count' : 'sum'})\
                                        .rename(columns = {'listening_count' : 'listening_count_sum'})\
                                        .sort_values(by=['listening_count_sum'], ascending=False)\
                                        .reset_index()\
                                        .head(100000)
#df_filter_1 



## first 1e+4 "extreme users" who can listen to a single song more than a thousand times

df_filter_2 = df_triplets_metadata_spotify.groupby('user')\
                                        .agg({'listening_count' : 'sum'})\
                                        .rename(columns = {'listening_count' : 'listening_count_sum'})\
                                        .sort_values(by=['listening_count_sum'], ascending=False)\
                                        .reset_index()\
                                        .head(10000)


## first 1e+4 users who listen multiple songs
'''
df_filter_2 = df_triplets_metadata_spotify.groupby('user')\
                                        .agg({'listening_count' : 'sum', 'listening_count' :'count'})\
                                        .rename(columns = {'listening_count' : 'listening_count_sum'})\
                                        .sort_values(by=['listening_count_sum'], ascending=False)\
                                        .head(10000)
'''

df_filter_2.head(11)
```

```python
df_triplets_metadata_spotify_sample1 = df_triplets_metadata_spotify[df_triplets_metadata_spotify["user"].isin(df_filter_1['user'])]\
                                                                                                        .sort_values(by = 'user')
#df_triplets_metadata_spotify_sample1

df_triplets_metadata_spotify_sample2 = df_triplets_metadata_spotify[df_triplets_metadata_spotify["user"].isin(df_filter_2['user'])]\
                                                                                                        .sort_values(by = 'user')
df_triplets_metadata_spotify_sample1
```
