```python
import os
from pathlib import Path
import pandas as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import ipynb
```

```python
from ipynb.fs.full.datamanagement import user_filter
```

```python
# Read the data
CUR_DIR = os.path.abspath('')

DATA_DIR = Path(CUR_DIR).parent / "data"
FILE_PATH = DATA_DIR/'triplets_metadata_spotify.csv'
df = pd.read_csv(FILE_PATH, index_col = 0)
```

```python
df.head()
```

```python
# Compute the number of listening for each pairs user/songs regardless the release of the song
user_listenings = df.groupby(['user', 'song_name', 'artist_name']).agg({'listening_count' : sum}).reset_index()
user_listenings.head()
```

```python
# Get the sum of the listering count for a user and the number of songs he have listened
users_summary = user_listenings.groupby('user').agg({'listening_count' : sum, 'song_name' : len}).reset_index()
users_summary.columns = ['user', 'listening_count', 'track_count']
```

```python
users_summary.head()
```

```python
users_summary['avg_listening_count_by_song'] = users_summary['listening_count']/users_summary['track_count']
```

```python
users_summary.describe()
```

```python
print('Percent of users who have one listening : %.2f%%' %(sum(users_summary.listening_count == 1)*100/len(users_summary)))
print('Number of users who have less than 5 listening :', sum(users_summary.listening_count > 5))
print('Percent of users who have listened more than 5 different songs : %.2f%%' %(sum(users_summary.track_count > 5)*100/len(users_summary)))
print('Number of users who have listened more than 5 different songs :', sum(users_summary.track_count > 5))
```
